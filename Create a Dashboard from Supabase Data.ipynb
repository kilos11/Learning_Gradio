##Create a small Gradio Dashboard with 2 Barplots that plots the prices and inventories of all of the items every minute and updates in real-time:

```python
!pip install gradio
```

```python
import gradio as gr

with gr.Blocks() as dashboard:
    with gr.Row():
        gr.BarPlot(
            read_data,
            x="product_id",
            y="price",
            title="Prices",
            every=60

        )
        gr.BarPlot(
            read_data,
            x="product_id",
            y="inventory_count",
            title="Inventory",
            every=60
        )

dashboard.queue().launch()
```
